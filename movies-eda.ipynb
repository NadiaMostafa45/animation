```python
import pandas as pd

import seaborn as sns

# Load the dataset
df = pd.read_csv('/kaggle/input/imdb-movie-insights/IMDB_Movie_Insights.csv')
df
```

```python
# Display the column names in the DataFrame
print(df.columns)
```

```python
import matplotlib.pyplot as plt
sns.histplot(df['Age'], bins=30, kde=True)
plt.title('Distribution of Age')
plt.show()
```

```python
sns.histplot(df['Duration(minutes)'], bins=30, kde=True)
plt.title('Distribution of Duration(minutes)')
plt.show()
```

```python
sns.scatterplot(x='Ratings', y='Gross Profit($M)', data=df)
plt.title('Rating vs Gross Profit')
plt.show()
```

```python
sns.histplot(df['Votes'], bins=30, kde=True)
plt.title('Distribution of Votes')
plt.show()
```

```python
sns.pairplot(df[['movie_name', 'Year', 'Age', 'Duration(minutes)', 'Ratings',
       'Gross Profit($M)', 'Votes']])
plt.show()
```

```python
sns.histplot(df['Year'], bins=30, kde=True)
plt.title('Distribution of Year')
plt.show()
```
